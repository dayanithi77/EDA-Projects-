```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
df=pd.read_csv("C:/Users/DAYANITHI/Downloads/datasets/tracks.csv")
df.head(1)
df.shape
```

```python
top_10=df.query('popularity > 90',inplace=False).sort_values('popularity', ascending=False)
top_10[:10] #gives the first 10 most popular alone 
# .query is used in pandas to query a dataframe , sort values used to sort in ascending or descending order 
```

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index) 
df.head()
```

```python
df[['artists']].iloc[18] #to get a specific data point / use iloc .
```

```python
df ["duration"]= df["duration_ms"].apply(lambda x: round (x/1000))  # to covert duration from ms to normal seconds .
df.drop("duration_ms", inplace=True, axis=1)   # then dropping the previous duration in ms . 
```

```python
corrdf=df.drop(['explicit','key','mode'],axis=1).corr(method='pearson') #dropping these three cus they dont contriute much 
plt.figure(figsize=(14,6))  #set fig size before the plot , if after it wo'nt apply .
heat=sns.heatmap(corrdf,annot=True,vmin=1,vmax=-1,linewidths=1)
heat.set_title("correlation plot")
heat.set_xticklabels(heat.get_xticklabels(),rotation=90)# xticks with rotation will make the column names visible at 90 degree , if set at 180 they will clash 
```

```python
samdf=df.sample(int(0.004*len(df)))
print(len(samdf))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=samdf,y="loudness",x="energy",color='c').set(title="loudness VS Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=samdf,y="popularity",x="acousticness",color='b').set(title="popularity VS acousticness")
```

```python
df['dates']=df.index.get_level_values("release_date")
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,kind='hist')
```

```python
total_dr = df.duration
fig_dims=(18,7)
fig,ax = plt.subplots (figsize = fig_dims) 
fig = sns.barplot (x = years, y =total_dr,ax = ax,errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=98)
```

```python
total_dr=df.duration 
sns.set_style(style="whitegrid")
fig_dims=(10, 5)
fig, ax = plt.subplots (figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration") 
plt.xticks (rotation=60)
```
